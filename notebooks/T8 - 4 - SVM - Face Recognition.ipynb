```python
from sklearn.datasets import fetch_lfw_people
import matplotlib.pyplot as plt
```

```python
faces = fetch_lfw_people(min_faces_per_person= 40) # entre menos descarga mas
```

```python
print(faces.target_names)
```

```python
print(faces.images.shape)
print(faces.images.shape[0]," caras")
```

```python
fig, ax = plt.subplots(5,8, figsize = (16,9))
for i, ax_i in enumerate(ax.flat):
    ax_i.imshow(faces.images[i], cmap="bone")
    ax_i.set(xticks=[],yticks=[],xlabel=faces.target_names[faces.target[i]])
```

```python
62*47 ## 62 pixeles de altura y 47 de anchura = 2914 pixeles
```

### si se usara cada pixel como un rasgo tendriamos 2914 variables 2914 dimensiones
### No es viable por eso se usa un preprocesado par sacar los rasgos mas importantes

```python
from sklearn.svm import SVC
from sklearn.decomposition import RandomizedPCA #principal component analized
from sklearn.pipeline import make_pipeline #tuberia permite unir dos tareas randomized  seguido de SVC
```

```python
pca = RandomizedPCA(n_components = 150, whiten = True, random_state=42)# n_components = numero de variables importantes
svc = SVC(kernel="rbf", class_weight="balanced") # por que podria tener mas sentido buscar agrupaciones circulares que lineales
model = make_pipeline(pca,svc) # creo un modelo que primero obtiene los rasgos principales y luego el SVC
```

```python
from sklearn.cross_validation import train_test_split
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(faces.data,faces.target,random_state = 42)
```

```python
from sklearn.grid_search import GridSearchCV #Busqueda en parrilla
#para la validacion cruzada
```

```python
#Con esto buscamos la C optima la gamma que se usa para el calculo radial que usa el model rbf
param_grid = {
    "svc__C":[0.1,1,5,10,50],
    "svc__gamma":[0.000,0.0005,0.001,0.005,0.01]
}
grid = GridSearchCV(model, param_grid) # busca en la parrilla con el modelo 

%time grid.fit(Xtrain, Ytrain)# vemos el tiempo que tarda casi 2 minuto
```

```python
print("Los parametros optimos de C y gama son :",grid.best_params_)
```

```python
classifier = grid.best_estimator_ #obtendriamos el model con mejor estimacion 
yfit = classifier.predict(Xtest) #y hacemos la prediccion
yfit.shape
```

## Notas
 * El valor obtenido no debe estar en las orillas, si aparece en las orillas expandir parillas con mas valores
 * Cuando este en medio repetir el experimento con valores cercanos al obtenido y ver si hay uno que sea todavia mejor

```python
#Repetimos experimento
param_grid = {
    "svc__C":[1,2,3,4,5,6,7],
    "svc__gamma":[0.0008,0.0009,0.001,0.002,0.003,0.004,0.005]
}
grid = GridSearchCV(model, param_grid)
grid.fit(Xtrain, Ytrain)
```

```python
print("Los parametros optimos de C y gama son :",grid.best_params_)
```

```python
classifier = grid.best_estimator_ #obtendriamos el model con mejor estimacion 
yfit = classifier.predict(Xtest) #y hacemos la prediccion
yfit.shape
```

```python
fig, ax = plt.subplots(6,10,figsize=(16,9))
for i, ax_i in enumerate(ax.flat): #eje y posicion
    ax_i.imshow(Xtest[i].reshape(62,47), cmap="bone")# bone: blanco y negro
    ax_i.set(xticks=[],yticks =[])
    ax_i.set_ylabel(faces.target_names[yfit[i]].split()[-1] ,color="black" if yfit[i] == Ytest[i] else "red" )
fig.suptitle("Prediccion de las imagenes (incorrectas)")
```

## Nota con una C = 5 predice de mejor manera aunque el optimo del segundo experimento donde C = 2

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Ytest, yfit, target_names = faces.target_names))
#                           prueba , predecido, nombre
# precision : especisidad
# recall : especisidad 1 - falso negativos
# support cantidad de veces que aparece en el dataset
```

```python
from sklearn.metrics import confusion_matrix
```

```python
mat = confusion_matrix(Ytest, yfit)
mat #como se ve fea por eso es mejor continuar
```
