## Suppor Vector Machines

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns; sns.set()
```

```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
X, Y = make_blobs(n_samples=50, centers = 2, random_state = 0, cluster_std=0.6)
```

```python
plt.scatter(X[:,0],X[:,1], c = Y, s = 50, cmap="autumn")
```

```python
xx = np.linspace(-1,3.5)
plt.scatter(X[:,0],X[:,1], c = Y, s = 50, cmap="autumn")
plt.plot([0.5],[2.1], 'x', color="blue", markeredgewidth=2, markersize=10)

for a, b in [(1,0.65),(0.5,1.6),(-0.2,2.9)]:
    yy = a * xx + b
    plt.plot(xx,yy,"-k")
plt.xlim(-1,3.5)
```

### En este grafico vemos como si la recta divisora es la primera o la ultima si se puede clasificar el punto azul pero si es la de enmedio entonces no tenemos una forma clara  y a esto hay que agregar que problamente tengamos mas rectas divisoras


## Maximizacion del margen

```python
xx = np.linspace(-1,3.5)
plt.scatter(X[:,0],X[:,1], c= Y, s = 50, cmap = "autumn")
plt.plot([0.5],[2.1], 'x', color = 'blue', markeredgewidth=2, markersize = 10)


for a, b, d in [(1,0.65,0.33),(0.5,1.6,0.55,),(-0.2,2.9,0.2)]:
    yy = a * xx + b
    plt.plot(xx,yy,"-k")
    plt.fill_between(xx, (yy - d)  , (yy+d) , edgecolor='none', color="#BBBBBB",alpha=0.4)
    
plt.xlim(-1,3.5)
```

## Creacion del modelo SVM

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel="linear", C= 1E10)
model.fit(X,Y)
```

```python
def plt_svc(model, ax = None, plot_support = True):
    """Plot de la función de decission para una clasificacion en 2D con SVC """
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    ## Generamos la parrilla de puntos para evaluar el modelo
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(yy,xx)
    xy =  np.vstack([X.ravel(), Y.ravel()]).T #T genera una tupla
    P = model.decision_function(xy).reshape(X.shape) # recorta los datos para que tengan la dimension apropiada
    
    ## Representamos las fronteras y los margenes del SVC
    ax.contour (X,Y,P,colors="k",levels=[-1,0,1],alpha = 0.5, linestyles = ["--","-","--"])
    print(model.support_vectors_)
    
    if(plot_support):
        ax.scatter(model.support_vectors_[:,0],
                  model.support_vectors_[:,1],
                  s = 300, linewidth=1, facecolors="black")
    
    ax.set_xlim(xlim)# por seguridad de que los limites no se vean afectados
    ax.set_ylim(ylim)# por seguridad de que los limites no se vean afectados
```

```python
plt.scatter(X[:,0],X[:,1], c = Y, s = 50, cmap="autumn")
plt_svc(model,plot_support=True)

# los puntos negros marcan los soportes de soporte
# lo vectores de soporte marcan los puntos donde el pasillo choca con los grupos de puntos mas cercanos al hiperplano
# serian las coordenas de los puntos estos se llaman pivotes
# Cualquier punto mas alla de los pivotes se puede clasificar sin problema
```

```python
def plot_svm(N=10,ax=None,n_samples = 200):
    X, Y = make_blobs(n_samples = n_samples,centers=2,random_state=0,cluster_std=0.6)
    
    X = X[:N]
    Y = Y[:N]
    
    model = SVC(kernel = "linear", C = 1E10)
    model.fit(X,Y)
    ax = ax or plt.gca()
    ax.scatter(X[:,0], X[:,1], c = Y, s= 50, cmap="autumn")
    ax.set_xlim(-1,4)
    ax.set_ylim(-1,6)
    plt_svc(model,ax)
```

```python
fig, ax = plt.subplots(1,2, figsize=(16,6))# 1 fila 2 columnas
fig.subplots_adjust(left=0.0625, right = 0.95, wspace = 0.1)#Agregar margenes en porcentaje al tamaño total 0 - 1
for ax_i , N, in zip(ax,[60,120]):
    plot_svm(N,ax_i)
    ax_i.set_title("N={0}".format(N))
```

## La funcion plot_svm

Esta funcion crea una nube de puntos aleatoria que al tener el mismo random_state seria una nube aleatoria pero repetitiva

Esto con la intencion de que el modelo primero calcule los soportes vectoriales con 60 y luego con 120 para ver que no importa el numero de puntos al final el pasillo no cambia

Pero si agrandamos la muestro con un demo vemos al mismo resultado ahora con 2000 puntos

```python
fig, ax = plt.subplots(1,3, figsize=(16,6))# 1 fila 2 columnas
fig.subplots_adjust(left=0.0625, right = 0.95, wspace = 0.1)#Agregar margenes en porcentaje al tamaño total 0 - 1
for ax_i , N, in zip(ax,[600,1200,2000]):
    plot_svm(N,ax_i,2000)
    ax_i.set_title("N={0}".format(N))
```
