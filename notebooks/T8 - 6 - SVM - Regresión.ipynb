## SVM para Regresi√≥n

```python
import numpy as np
import matplotlib.pyplot as plt
```

Se pueden usar matrices que generarian graficas de n dimensiones

Entonces para la representacion grafica es bueno usar solo 1 variable de entrada

Por eso para este ejemplo solo se usa 1 arreglo

```python
length = 200
random = 5
l_rand = int(length/random)
X = np.sort(5* np.random.rand(length,1),axis = 0)
Y = np.cos(X).ravel()
Y[::random] += 3*(0.5 - np.random.rand(l_rand))# esto hace que cada 5 numero de agregue una desviacion
```

```python
plt.scatter(X,Y, color="darkorange",label="data")
```

```python
from sklearn.svm import SVR
```

```python
C = 1e3
svr_lin = SVR(kernel="linear", C=C)
svr_rbf = SVR(kernel="rbf", C=C, gamma = 0.1)
svr_pol = SVR(kernel="poly", C=C, degree = 3)
```

```python
y_lin = svr_lin.fit(X,Y).predict(X)
y_rbf = svr_rbf.fit(X,Y).predict(X)
y_pol = svr_pol.fit(X,Y).predict(X)
```

```python
lw = 2 #line weigth 
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X,Y,color="darkorange",label="data")
plt.plot(X,y_lin, color="navy",lw=lw, label="SVM Lineal")
plt.plot(X,y_rbf, color="c",lw=lw, label="SVM RBF")
plt.plot(X,y_pol, color="cornflowerblue",label="SVM POLY")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Support Vecot Regression")
plt.legend()
plt.show()
```

### Los SVM tienen buenas cualidades como por ejemplo poder manejar datasets que tengan mas columnas que filas lo trabaja sin problemas
### Pero a medida que agregamos variables(columnas o dimensiones) el tiempo de entrenamiento se incrementa y afinar los hiperparametros podria llegar a ser mas costoso que usar otro algoritmo CUIDADO
